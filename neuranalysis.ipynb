```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import os

%matplotlib inline
```

```python
vertices = np.empty([0, 3])

with open('1_0001Full100.obj', 'r') as file:
    for line in file:
        if line.startswith('v'):
            # e.g. 'v 192.2170 264.2010 76.3423 \n'
            coordinates = line.rstrip()
            coordinates = coordinates.split(' ')[1:]
            x, y, z = [float(i) for i in coordinates]
            vertices = np.append(vertices, [[x, y, z]], axis=0)

# Split out coordinates for ease of graphing with matplotlib's scatter3D
xs = vertices[:,0]
ys = vertices[:,1]
zs = vertices[:,2]
```

```python
distance = np.sqrt(xs**2 + ys**2 + zs**2)
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(xs, ys, zs, c=distance, cmap='viridis')
```
